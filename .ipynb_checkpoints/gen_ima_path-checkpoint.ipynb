```python
from PIL import Image
import glob
import os
from matplotlib import pyplot as plt
```

```python
files = glob.glob(os.path.join("Pf","*.png"))
print(files)
```

```python
img = Image.open(files[0])
plt.imshow(img)
```

```python
croped = img.crop((0,200,1920,1000))
plt.imshow(croped)
```

```python
count = 1
for file in files:
    print('cropped/{:010d}.png'.format(count))
    img = Image.open(file)
    croped = img.crop((0,200,1920,1000)).save('cropped/{:010d}.png'.format(count))
    count = count + 1
```
